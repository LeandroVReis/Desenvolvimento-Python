### Read

```python
import pyodbc

dados_conexao = ("Driver={SQLite3 ODBC Driver};"
                 "Server=localhost;"
                 "Database=chinook.db")

conexao = pyodbc.connect(dados_conexao)
print('Conexão bem sucedida!')
```

```python
cursor = conexao.cursor()

cursor.execute("""
    SELECT * FROM customers
""")

valores = cursor.fetchall()
descricao = cursor.description

cursor.close()
```

```python
import pandas as pd

colunas = [tupla[0] for tupla in descricao]
print(colunas)
```

```python
tabela_customers = pd.DataFrame.from_records(valores, columns=colunas)
display(tabela_customers[:5])
```

```python
tabela_clientes = pd.DataFrame.from_records(valores, columns=['ID', 'NOME', 'SOBRENOME', 'COMPANHIA', 'ENDEREÇO', 'CIDADE', 'ESTADO', 'PAÍS', 'CEP', 'TELEFONE', 'FAX', 'EMAIL', 'SUPPORTREPID'])
display(tabela_clientes[:5])
```

```python
conexao.close()
print('Conexão finalizada!')
```
